```python
import SimpleITK as sitk
```

```python
from myshow import myshow, myshow3d
```

```python
img1 = sitk.ReadImage("/Users/tmstout/Dropbox/CT_Images/CT1.nii.gz")
```

```python
grad = sitk.GradientMagnitude(img1)
```

```python
myshow3d(grad)
```

```python
thresh = sitk.OtsuThreshold(grad, insideValue = 0, outsideValue = 1 )
```

```python
myshow3d(thresh)
```

```python
canny = sitk.CannyEdgeDetection(grad,lowerThreshold=0, upperThreshold=1)
```

```python
myshow3d(canny)
```
